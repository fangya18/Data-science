#### Ex.1 
$ C_{N}_{M}=\frac{M!}{N!(M-N)!}$


```python
def comb(m,n):
    fm=1 
    for num in range(1,m+1):
        fm *=num
    fn=1
    for num in range(1,n+1):
        fn*=num
    fm_n=1
    for num in range (1, m-n+1):
        fm_n=num
    print(fm//(fn*fm_n))
```

```python
comb(4,2)
```

#### EX2. Factorial

```python
def fac(x):
    result=1
    for i in range(1, x+1):
        result *= i
    return result
```

```python
fac(4)
```

#### EX3. Roll Dice

```python
from random import randint

def roll_dice(n):
    total=0
    for _ in range (n):
        total += randint(1,6)
    return total
```

```python
roll_dice(3)
```

#### EX4. Sum Function: definite number VS freesytle adding

```python
### a certain number of input
def add(a,b,c):
    return a+b+c
print(add(3,2,0))
```

```python
### Freestyle adding
def add1(*args):
    sum=0
    for val in args:
        sum +=val
    return sum
```

```python
add1(1,2)
```

```python
add1(1,2,3,4)
```

```python
def gcd(x,y):
    if x >y:
        x,y=y,x
    for factor in range (x,0,-1):
        if x % factor==0 and y % factor ==0 :
            return factor
    
```

```python
gcd (4,3)
```

```python
gcd (4,8)
```

```python
def lcm(x,y):
    return x*y//gcd(x,y)
```

```python
lcm(4,8)
```

#### Ex5. Palindrome 

```python
def pali(x):
    temp=x
    total=0
    while temp >0:
        total=total*10 +temp%10
        temp //=10
    return total==x
```

```python
pali(123)
```

```python
pali(12321)
```

#### EX6. is it Prime number ?

```python
import  math
def is_prime(x):
    for factor in range (2, int(math.sqrt(x))+1):
        if x % factor ==0:
            return False
        else:
            return True
```

```python
is_prime(4)
```

```python
is_prime(12321)
```

```python
is_prime(5)
```
