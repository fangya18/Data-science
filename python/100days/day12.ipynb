#### Ex1. Bilibili ID Number Validation

* Username is consisted by alphabet, numbers or underscore with in 6~20 letters
* ID number is between 6 - 9 and not start with 0



```python
from IPython.display import Image
Image(filename="C:/Users/fatan/Desktop/ALL/1Study/6.Wordpress/sas/water.PNG", width=400, height=200)
```

```python
import re

def bilibili(name, id_no):
    m1=re.match(r'^[0-9a-zA-Z_]{6,20}$', name )
    if not m1 :
        print ("Please enter a valid username")
    m2=re.match(r'^[1-9]\d{5,8}', id_no)
    if not m2:
        print("Please enter valid bilibili ID")
    if m1 and m2:
        print("Welcome back in Bibibili! ⭐ 一键三连哦！")
    
```

```python
bilibili("xianrenwangyiheng", "487435226")
```

```python
bilibili("yibowang", "0805")
```

```python
import re

def nyc_tel():
    nyc=re.compile (r'(?<=\D)(917|212)\d{7}(?=\D)')
    conversation="""
    Y: hi, can i talk to L?
    B: Sorry, this is B.
    Y: is your number 9175289549?
    B: No, my number is 289012324.
    Y: sorry, 886
    B: How could you mix 917xx with 289xxx ?
    """
    mylist=re.findall(nyc,conversation)
    print(mylist)
    #print("-----⭐------")
    #for temp in nyc.finditer(conversation):
    #    print(temp.group())
    print("⭐ NYC #----")
    m= nyc.search(conversation)
    while m:
        print(m.group())
        m=nyc.search(conversation, m.end())
    
```

```python
import re

def change():
    sentence = '操 ！ 喜真他妈的帅的一逼， 太屌了. the dance show if fucking amazing ！'
    purified = re.sub('操|fucking|他妈的|屌',
                      'awsl', sentence, flags=re.IGNORECASE)
    print(purified)  # 你丫是*吗? 我*你大爷的. * you.
```

#### Ex4. Split Long Strings
* Assign special characters to split a long string

```python
import re

def split_str():
    poem= "Take derivates with Sadness, Take the integral with Happiness, We may forget the L'Hopital Rule. Dear friends, dont worry! Remember we still have the Taylor Expansion."
    sentence_list=re.split(r'[,.!.]', poem)
    while " " in sentence_list:
        sentence_list.remove(" ")
    print(sentence_list)
    
    
```

```python
split_str()
```
