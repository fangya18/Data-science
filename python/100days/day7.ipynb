#### Ex6. Fibonacci number

```python
def fib(n):
    f1,f2=0,1
    for i in range (1,n):
        f1,f2=f2,f1+f2
        yield f1
    
def fib_seq(n):
    for val in fib(n):
        print(val)
        
```

```python
fib(10)
fib_seq(10)
```

#### Ex11. Generate a random password

```python
import random

def generate_code(n):
    all_chars="01234abcdABCD"
    # a code with random selection
    last_pos=len(all_chars) -1
    code=" "
    for i in range(n):
        index=random.randint(0,last_pos)
        code += all_chars[index]
    return code
```

```python
generate_code(4)
```

#### Ex12. Get the suffix

```python
def suffix(filename, has_dot=False):
    pos=filename.rfind(".")
    if 0< pos <len(filename) -1:
        index= pos if has_dot else pos +1
        return filename [index:]
    else:
        return " "
```

```python
suffix("yibo.png")
```

```python
suffix("yibo.gif")
```

#### Ex13. Find the largest and second largest element

```python
def max2(x):
    m1,m2=(x[0], x[1]) if x[0] >x[1] else (x[1],x[0])
    for index in range (2, len(x)):
        if x[index] >m1:
            m2=m1
            m1= x[index]
        elif x[index] >m2:
            m2=x[index]
    return m1,m2
```

```python
max2([1,2,3,4,3,8])
```

#### Ex14. Count the number of the day in a year

```python
def leap(year):
    return year %4==0 and year %100 != 0 or year %400 ==0

def day_count(year, month, day):
    days_of_months=[
        [31,28,31,30,31,30,31,31,30,31,30,31],
        [31,29,31,30,31,30,31,31,30,31,30,31]
    ][leap(year)]
    total=0
    for index in range(month - 1):
        total += days_of_months[index]
    return total+day

def count():
    print(day_count(1949,10,1))
    print(day_count(2020,8,13))
```

```python
count()
```

#### Ex15. Pascal Triangle

```python
def pascal(n):
    tri=[[]]*n
    for row in range(len(tri)):
        tri[row]=[None]*(row+1)
        for col in range(len(tri[row])):
            if col==0 or col==row:
                tri[row][col]=1
            else:
                tri[row][col]=tri[row-1][col]+tri[row-1][col-1]
            print(tri[row][col], end="\t")
        print()
```

```python
pascal(5)
```

#### Ex16. Lottery 

```python
from random import randrange, randint, sample
def lottery():
    # random number in 1 to 35
    balls=[x for x in range(1,36)]
    select=[]
    select=sample(balls,6)
    select.sort()
    # a special number
    select.append(randint(1,16))
    return select

def buy(n):
    for i in range(n):
        display(lottery())
```

#### Ex17. Josephus permutation
There are n people standing in a circle waiting to be executed. The counting out begins at some point in the circle and proceeds around the circle in a fixed direction.

```python
def josephus():
    persons=[True]*30
    counter, index, number=0,0,0
    while counter <15:
        if persons[index]:
            number +=1
            if number==9:
                persons[index]= False
                counter +=1
                number=0
        index +=1
        index %=30
    for person in persons:
        print("Lucky" if person else "-_-", end=" ")
```

```python
josephus()
```

#### Tit Tac Toe
Lyo: 四子不算赢，只有双三才算！我从来不靠别人看不到来赢。
Only double three counted win for me
\
落子无悔！ 
后悔也要假装不后悔

```python
import os


def print_board(board):
    print(board['TL'] + '|' + board['TM'] + '|' + board['TR'])
    print('-+-+-')
    print(board['ML'] + '|' + board['MM'] + '|' + board['MR'])
    print('-+-+-')
    print(board['BL'] + '|' + board['BM'] + '|' + board['BR'])


def tit_tac_toe():
    init_board = {
        'TL': ' ', 'TM': ' ', 'TR': ' ',
        'ML': ' ', 'MM': ' ', 'MR': ' ',
        'BL': ' ', 'BM': ' ', 'BR': ' '
    }
    begin = True
    while begin:
        curr_board = init_board.copy()
        begin = False
        turn = 'x'
        counter = 0
        os.system('clear')
        print_board(curr_board)
        while counter < 9:
            move = input('%s Please make a move: ' % turn)
            if curr_board[move] == ' ':
                counter += 1
                curr_board[move] = turn
                if turn == 'x':
                    turn = 'o'
                else:
                    turn = 'x'
            os.system('clear')
            print_board(curr_board)
        choice = input('One more?(yes|no)')
        begin = choice == 'yes'
```

```python
tit_tac_toe()
```
