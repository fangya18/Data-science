#### Ex1. Narcissistic number



##### numbers that can be represented as the sum of the cube of each digits. 

$$153= 1^{3} + 5^{3} +3^{3}$$

```python
def nar():
    for x in range (1,1001):
        s=x%10
        t=x//10%10
        h=x//100
        if x==s**3+t**3+h**3:
            print("%d is a narcissistic number" %x)
           
```

```python
nar()
```

#### Ex2. Reversible the Number



```python
def rev(x):
    reverse=0
    while x>0:
        reverse= reverse*10 + x%10 
        x //=10
    print(reverse)
```

```python
rev(12)
```

```python
rev(12345)
```

#### Ex3. 100 dollars with Chickens



how many chicks, rooster, and hen we can purchase with 100$ ?
rooster 5 dollar each
hen 3 dollar each
chicks 1 dollar for three

```python
def buy_chicken():
    for r in range (0,20):
        for h in range (0,33):
            c =100 -r-h
            if r*5+ h*3+ c/3==100:
                print("100 dollars can buy %d rooster , %d hen, %d chicks" %(r,h,c))
```

```python
buy_chicken()
```

#### Ex4. CRAPS GAME
![](https://encrypted-tbn0.gstatic.com/images?q=tbn%3AANd9GcQY207ObpLkpX4OBw1sNEujI4eovWkF-T0i9rYAscsI0_0DnKA7iGVHYoJ01Zuk438VqyLDYS2B&usqp=CAc)

* player 1 win - dice 7 or 11
* player 2 win - dice 2 , 3 or 12
* player 2 win if player 1 roll 7
* player 1 win if player 1 roll the previous dice number  

--- continue until wallet is empty


```python
from random import randint

def craps(money):
    while money>0:
        go_on=False
        while True:
            bet= int(input())
            if 0<bet <= money:
                break
        first= randint(1,6) +randint(1,6)
        print("Player 1 has %d" %first)
        if first ==7 or first==11:
            print("Player 1 Won ! 撒花 **")
            money+= bet
        elif first==2 or first==3 or first==12:
            print("Player 2 Won ! -_-!!!")
            money-= bet
        else:
            go_on=True
        while go_on:
            go_on=False
            current=randint(1,6) +randint(1,6)
            print("Player 1 has %d" %current)
            if current ==7:
                print("Player 2 won -_-!")
                money-=bet
            elif current==first:
                print("Player 1 won! 撒花**")
                money+=bet
            else:
                go_on=True
    print("Save some money to buy Ice-cream !")           
            
```

```python
craps(5)
```

#### Ex5. Fibonacci Sequence & 姑苏小白兔！
$$ f_{n+2}=f_{n}+f_{n+1}$$


```python
def fibo(m):
    f1=0
    f2=1
    for _ in range (1,m):
        f1,f2=f2, f1+f2
        print (f1, end=" ")
    print ("云深不知处，姑苏小白兔 Cute ^_^!")
```

```python
fibo(20)
```

#### Ex6. Perfect Number


the number is equal to its sum of proper divisors 
$$ 6 = 1+ 2+ 3$$

```python
from math import sqrt
def perfect(m):
    for i in range (1,m):
        s=0
        for factor in range (1, int(sqrt(i))+1):
            if i % factor==0:
                s +=factor
                if factor >1 and i//factor != factor:
                    s+= i//factor
        if s==i:
            print(i,"is a perfect number" )
    print("So you are perfect2!")
            
```

```python
perfect(10000)
```

#### Ex7. Prime Number Finder


The only divisor of the number is 1 and itself

```python
from math import sqrt
def prime(m):
    for i in range (2,m):
        isprime=True
        for factor in range (2, int(sqrt(i))+1):
            if i % factor ==0:
                isprime=False
                break
        if isprime:
            print(i)
```

```python
prime(100)
```
