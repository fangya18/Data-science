#### Ex2. Write Text File
Print out python result and save it as text files on desktop.
Find the Prime numbers 
* Save prime numbers $<10$ as prime10.txt
* Save prime numbers $<100$ as prime100.txt
* Save prime numbers $<200$ as prime200.txt

```python
from math import sqrt

def is_prime(n):
    assert n > 0
    for factor in range(2, int(sqrt(n)) + 1):
        if n % factor == 0:
            return False
    return True if n != 1 else False


def p_prime():
    filenames = ('prime10.txt', 'prime100.txt', 'prime200.txt')
    fs_list = []
    try:
        for filename in filenames:
            fs_list.append(open(filename, 'w', encoding='utf-8'))
        for number in range(1, 200):
            if is_prime(number):
                if number < 10:
                    fs_list[0].write(str(number) + '\n')
                elif number < 100:
                    fs_list[1].write(str(number) + '\n')
                else:
                    fs_list[2].write(str(number) + '\n')
    except IOError as ex:
        print(ex)
        print('Opps, Try again @_@!')
    finally:
        for fs in fs_list:
            fs.close()
    print('Done ^_^! ')
```

```python
p_prime()
```

*wb : write as binary

```python
from IPython.display import Image
Image(filename="C:/Users/fatan/Desktop/binary.jpg")
```

```python
def r_binary():
    try:
        with open("C:/Users/fatan/Desktop/binary.jpg" ,"rb") as fs1:
            data=fs1.read()
            print(type(data))
        with open("C:/Users/fatan/Desktop/b2.jpg" ,"wb") as fs2:
            fs2.write(data)
    except FileNotFoundError as e:
        print('No file existed -_-|||')
    except IOError as e:
        print('Something is wrong @_@') 
    print("Done! ~_~")
```
