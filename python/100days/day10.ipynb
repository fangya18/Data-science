#### Ex1. Make a Message box

```python
import tkinter
import tkinter.messagebox

def box():
    flag=True
    
    def change_label_text():
        nonlocal flag
        flag=not flag
        color, msg=("red", "Dance with 易燃装置 jackson Yee!")\
            if flag else("green", "Dance with 一波王炸 yibo Wang!")
        label.config(text=msg, fg=color)
    
    def confirm_2_quit():
        if tkinter.messagebox.askokcancel("Are you sure you dont want to dance?"):
            top.quit()

    top = tkinter.Tk()
    # set the window size
    top.geometry('240x160')
    # set the game title
    top.title('Street Dance of China')
    # set the welcome name
    label = tkinter.Label(top, text='Hola Dance', font='Arial -32', fg='blue')
    label.pack(expand=1)
    # set a button holder
    panel = tkinter.Frame(top)
    # create a button to link the "command" variables 
    button1 = tkinter.Button(panel, text='Next', command=change_label_text)
    button1.pack(side='left')
    button2 = tkinter.Button(panel, text='Quit', command=confirm_2_quit)
    button2.pack(side='right')
    panel.pack(side='bottom')
    # start the loop
    tkinter.mainloop()
        
```

```python
box()
```
